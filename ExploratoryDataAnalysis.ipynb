DERIVING NEW FEATURES

```python

# Compute 'fueleconomy' based on the provided formula
cars['fueleconomy'] = (0.55 * cars['citympg']) + (0.45 * cars['highwaympg'])
```

```python
# Convert 'price' to integer type
cars['price'] = cars['price'].astype(int)

# Create a copy of the dataframe
temp = cars.copy()

# Compute the mean price for each car company and merge it back to the dataframe
company_price_mean = temp.groupby(['CompanyName'])['price'].mean().reset_index()
company_price_mean.columns = ['CompanyName', 'AvgPrice']
temp = temp.merge(company_price_mean, on='CompanyName', how='left')

# Define bins and labels for binning
bins = [0, 10000, 20000, 40000]
labels = ['Budget', 'Medium', 'Highend']

# Create a new column 'carsrange' based on the binned 'AvgPrice'
temp['carsrange'] = pd.cut(temp['AvgPrice'], bins=bins, labels=labels, right=False)

# Update the original 'cars' dataframe with the binned categories
cars['carsrange'] = temp['carsrange']

# Display the first few rows of the updated dataframe
print(cars.head())
```
